1. Import thư viện & dữ liệu

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cấu hình hiển thị
sns.set(style="whitegrid")

# Đọc dữ liệu
columns = [
    "Pregnancies", "Glucose", "BloodPressure", "SkinThickness",
    "Insulin", "BMI", "DiabetesPedigreeFunction", "Age", "Outcome"
]
data = pd.read_csv("data/pima-indians-diabetes.data.csv", header=None, names=columns)

# Thông tin cơ bản
print(data.shape)
print(data.info())
```

Thống kê cơ bản

```python
# Thống kê cơ bản
print(data.describe())

# Phân bố nhãn
print(data["Outcome"].value_counts(normalize=True))

sns.countplot(x="Outcome", data=data)
plt.title("Phân bố Outcome (0 = Không bệnh, 1 = Bệnh)")
plt.show()
```

Xác định giá trị thiếu và bất thường

Trong y khoa, các giá trị 0 ở Glucose, BloodPressure, SkinThickness, Insulin, BMI là bất hợp lý → coi như missing.

```python
cols_with_zero_invalid = ["Glucose","BloodPressure","SkinThickness","Insulin","BMI"]
zero_counts = (data[cols_with_zero_invalid] == 0).sum()
print("Số lượng giá trị 0 bất thường:\n", zero_counts)
```

Thay thế 0 bằng NaN để kiểm tra:

```python
data_nan = data.copy()
data_nan[cols_with_zero_invalid] = data_nan[cols_with_zero_invalid].replace(0, np.nan)

print("Missing values sau khi thay NaN:\n", data_nan.isnull().sum())
```

Kết luận: nhiều missing, đặc biệt Insulin (374 giá trị ~49%).

Xác định ngoại lệ (Outliers)

Dùng phương pháp IQR.

```python
outliers = {}
for col in data.columns[:-1]:  # bỏ Outcome
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outlier_count = ((data[col] < lower) | (data[col] > upper)).sum()
    outliers[col] = outlier_count

print("Số lượng outlier theo IQR:\n", outliers)
```

Phân tích đơn biến

```python
data.hist(bins=30, figsize=(12,10))
plt.suptitle("Histogram các biến")
plt.show()

# KDE theo Outcome
for col in ["Glucose","BMI","Age"]:
    plt.figure(figsize=(6,4))
    sns.kdeplot(data=data, x=col, hue="Outcome", fill=True)
    plt.title(f"Phân phối {col} theo Outcome")
    plt.show()
```

Phân tích đa biến


```python
sns.boxplot(x="Outcome", y="Glucose", data=data)
plt.title("Boxplot Glucose theo Outcome")
plt.show()

sns.violinplot(x="Outcome", y="BMI", data=data)
plt.title("BMI theo Outcome")
plt.show()

sns.scatterplot(x="Age", y="Glucose", hue="Outcome", data=data)
plt.title("Age vs Glucose theo Outcome")
plt.show()
```

Ma trận tương quan

```python
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Ma trận tương quan")
plt.show()
```
