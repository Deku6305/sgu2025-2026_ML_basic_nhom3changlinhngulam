# import thư viện cần thiết


```python
import numpy as np
import matplotlib.pyplot as plt
```

# Tạo dữ liệu giả định

```python
np.random.seed(42)
X_log = np.random.randn(100, 1)
y_log = (X_log + np.random.randn(100, 1) * 0.4 > 0.5).astype(int)
m_log = len(X_log)
```

# Khởi tạo

```python
w_log = np.random.randn(1, 1) # Hệ số w
b_log = np.random.randn(1, 1) # Bias b
lr_log = 0.05 # Learning Rate
n_iterations_log = 300
```

# hàm sigmoid


```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

# Tính toán


```python
print("\nLogistic Regression (Batch Gradient Descent):")
for i in range(n_iterations_log):
    # Tính z và Dự đoán (a)
    z = X_log.dot(w_log) + b_log
    a = sigmoid(z)
    
    # Tính toán sai số
    error = a - y_log
    
    # Tính Loss (Binary Cross-Entropy)
    epsilon = 1e-15
    a = np.clip(a, epsilon, 1 - epsilon)
    loss = -np.sum(y_log * np.log(a) + (1 - y_log) * np.log(1 - a)) / m_log
    
    # Tính Gradient
    grad_w_log = X_log.T.dot(error) / m_log
    grad_b_log = np.sum(error) / m_log
    
    # Cập nhật tham số
    w_log = w_log - lr_log * grad_w_log
    b_log = b_log - lr_log * grad_b_log
    
    if i % 50 == 0:
        print(f"Iteration {i}: Loss={loss:.4f} | w={w_log[0][0]:.4f}, b={b_log[0][0]:.4f}")

print(f"\nOptimal parameters after 300 iterations: w={w_log[0][0]:.4f}, b={b_log[0][0]:.4f}")
```
