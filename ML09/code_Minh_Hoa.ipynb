# code minh họa
$$
f(x) = (x + 5)^2
$$

### Import thư viện

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_vals = np.arange(-10, 0.001, 0.1)
y_vals = (x_vals + 5)**2
plt.plot(x_vals, y_vals)
plt.show()
```

### Đạo hàm hàm f(x)
$$
f(x)' = \frac{df}{dx} = x(x + 5)
$$

### Cập nhật Gradient DescentL
$$
x_i = x_{i-1} - p\frac{df}{dx} = x_{i-1} - p*2(x_{i-1} + 5)
$$

### Code minh họa:

```python
max_iters = 1000    
precision = 0.0001  

lr        = 0.1     
x         = 3       
for i in range(max_iters):
    print("Step %d: x = %f  --> f(x) = %f"%(i, x, (x+5)**2))
    dxy = 2 * (x + 5)
    f_pre = (x + 5) ** 2
    x = x - lr * (dxy)
    f_cur = (x + 5) ** 2
    if abs(f_cur - f_pre) < precision: 
        break
print("The local minimum occurs at x=%f with f(x) = %f."%(x, (x + 5)**2))
```
