# Xử lí dữ liệu


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import IPython.display as display
import math
```

### 1.2 Đọc tập dữ liệu 

```python
data_path  = "../data/pima-indians-diabetes.csv"
data_names = ["Pregnant","2Hr_Gluco_Concentrate","Blood_Pressure","Skin_Fold","2Hr_Isnulin","BMI","Predigy","Age","Class"]
df = pd.read_csv(data_path, names = data_names)
```

```python
df.describe()
```

### 1.3 Xử lí dữ liệu thiếu

```python
df["2Hr_Gluco_Concentrate"] = df["2Hr_Gluco_Concentrate"].replace(0, np.nan)
df["2Hr_Gluco_Concentrate"] = df["2Hr_Gluco_Concentrate"].fillna(df["2Hr_Gluco_Concentrate"].median())
df["Blood_Pressure"] = df["Blood_Pressure"].replace(0, np.nan)
df["Blood_Pressure"] = df.groupby("Class")["Blood_Pressure"].transform(lambda x: x.fillna(x.median()))
df["Skin_Fold"] = df["Skin_Fold"].replace(0, np.nan)
df["Skin_Fold"] = df.groupby("Class")["Skin_Fold"].transform(lambda x: x.fillna(x.median()))
df["Skin_Fold"] = df["Skin_Fold"].replace(0, np.nan)
df["Skin_Fold"] = df.groupby("Class")["Skin_Fold"].transform(lambda x: x.fillna(x.median()))
df["2Hr_Isnulin"] = df["2Hr_Isnulin"].replace(0, np.nan)
df["2Hr_Isnulin"] = df.groupby("Class")["2Hr_Isnulin"].transform(lambda x: x.fillna(x.median()))
df["BMI"] = df["BMI"].replace(0, np.nan)
df["BMI"] = df.groupby("Class")["BMI"].transform(lambda x: x.fillna(x.median()))
df.describe()
```

- Thực hiện xử lí dữ liệu thiếu theo trung vị của lớp output

### 1.4 Normalization & Standardlize

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from numpy import set_printoptions
# separate array into input and output components
Y = df["Class"]
X = df.drop("Class",axis=1)

scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX = scaler.fit_transform(X)
scaler = StandardScaler().fit(X)
rescaledX = scaler.transform(X)
set_printoptions(precision=3)
X.describe()
```

- Rescale các giá trị số về [0,1] và theo phân bố chuẩn

## 2. Baseline Models

### 2.1 Spilt Train-Valuate

```python
from sklearn.model_selection import train_test_split
seed = 7
validation_size = 0.2
X_train, X_validation, Y_train, Y_validation = train_test_split(X, Y, test_size=validation_size, random_state=seed)
```

### 2.2 K_Fold Cross validation

```python
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
#learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
names = ["KNN","RandomForest","SVC","AdaBoost"]
clf = [KNeighborsClassifier(n_neighbors = 13),
       RandomForestClassifier(n_estimators=13),
       SVC()
      ,AdaBoostClassifier()]
def model_fit():
    scoring = 'accuracy'
    best = {"model": "","score":0}
    for i,name in enumerate(names):
        score = cross_val_score(clf[i], X_train, Y_train, cv=k_fold, n_jobs=1, scoring=scoring)
        print("Score of Model",name,":",np.mean(score)*100)
        if(np.mean(score)*100>best["score"]):
            best["name"] = name
            best["model"] = clf[i]
            best["score"] = np.mean(score)*100
    return best
best = model_fit()
print("Best Model is:",best["name"],"with score:",best["score"])
```

### 2.3 Đánh giá khả năng dự đoán với dữ liệu chưa từng xuất hiện

```python
for model in clf:
    model.fit(X_train, Y_train)
    acc = model.score(X_validation, Y_validation)
    print("Score of Model(not tune)", model.__class__.__name__, ":", acc*100)
```

- Tuy KNN và AdaBoost có điểm đánh đánh giá cao nhất đối với bài toán tuy nhiên khi dự đoán đối với dữ liệu chưa có thì Adaboost và randomforest lại tốt nhất. Nên ta sẽ thực hiện tune đối với 2 model sau
